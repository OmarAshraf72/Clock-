```python
import cv2
import numpy as np
import math
import time
import matplotlib.pyplot as plt
from datetime import date
import datetime

```

```python
def array_to_tuple(arr):
    return tuple(arr.reshape(1, -1)[0])

image = np.zeros((1000, 1000, 3), dtype="uint8")
font = cv2.FONT_HERSHEY_SIMPLEX

image[:] = [(0, 0, 0)]
cv2.circle(image, (500, 500), 250, (0, 0, 255), 4)
for i in range(0,360,6):
        x = int(500+250*math.cos(math.radians(i)))
        y = int(500+250*math.sin(math.radians(i)))
        cv2.line(image,(x,y),(x,y),(0,255,255),4)

#cv2.rectangle(image, (0, 0), (550, 140), (0, 0, 0), -1)


while 1:
    
    image_original = image.copy()
    today = date.today()
    time = datetime.datetime.now()
   
    cv2.putText(image, 'Date : ', (0, 50), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, str(today), (105, 50), 1, 2, (250  , 150, 0),2 )
    cv2.putText(image, 'Time : ', (0, 90), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, (time.strftime("%I:%M:%S %p") ), (105, 90), 1, 2, (250  , 150, 0), 2)
    date_time_now = datetime.datetime.now()
    cv2.putText(image, '12', (475, 240), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '11', (330, 275), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '10', (240, 380), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '9', (220, 510), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '8', (250, 640), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '7', (340, 750), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '6', (490, 780), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '5', (630, 740), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '4', (730, 650), 1, 2, (250  , 150, 0), 2 ) 
    cv2.putText(image, '3', (760, 515), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '2', (725, 375), 1, 2, (250  , 150, 0), 2 )
    cv2.putText(image, '1', (630, 280), 1, 2, (250  , 150, 0), 2 )

    

    time_now = date_time_now.time()
    hour = math.fmod(time_now.hour, 12)
    minute = time_now.minute
    second = time_now.second

    second_ = math.fmod(second * 6 + 270, 360)
    minute_ = math.fmod(minute * 6 + 270, 360)
    hour_ = math.fmod((hour * 30) + (minute / 2) + 270, 360)

    
   
    second_x = int(500 + 250 * math.cos(math.radians(second_)))
    second_y = int(500 + 250 * math.sin(math.radians(second_)))
    cv2.line(image, (500, 500), (second_x, second_y), (0, 240, 240), 2)
    
    minute_x = int (500 + 200 * math.cos(math.radians(minute_)))
    minute_y = int (500 + 200 * math.sin(math.radians(minute_)))
    cv2.line(image, (500, 500), (minute_x, minute_y), (0 ,230,0), 4)
    
    hour_x = int (500 + 160 * math.cos(math.radians(hour_)))
    hour_y = int (500 + 160 * math.sin(math.radians(hour_)))
    cv2.line(image, (500, 500), (hour_x, hour_y), (72.9 , 59.6, 5.1), 6)
    
   
    cv2.circle(image, (500, 500), 10,  (50, 50, 50), -1)
    
    
    cv2.imshow("clock", image)
   
  
    image = image_original.copy()

  
    k=cv2.waitKey(1)
    if k==ord('q'):
        break

cv2.destroyAllWindows()
```
